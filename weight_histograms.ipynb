```python
import warnings
warnings.filterwarnings('ignore')

import os
import sys
import torch
from src import laion_clap
import matplotlib.pyplot as plt
```

```python
names = [
    "reweighting_6_1e4_new",
    "reweighting_5_1e4_new",
    "reweighting_4_1e4_new",
    "reweighting_6_1e5_new",
    "reweighting_1_1e5_new",
    "reweighting_2_1e5_new",
    "reweighting_3_1e5_new",
    "reweighting_4_1e5_new",
]
```

```python
for name in names:
    checkpoint_path = f"/fs/nexus-scratch/milis/848K/CLAP/logs/{name}/checkpoints/epoch_latest.pt"

    sys.stdout = open(os.devnull, "w")
    sys.stderr = open(os.devnull, "w") 
    model = laion_clap.CLAP_Module()
    model.load_ckpt(checkpoint_path)
    model.eval()

    log_reweighting = model.model.text_branch.log_reweighting

    plt.figure()
    plt.hist(torch.exp(log_reweighting).detach().numpy(), bins=100, color='blue', alpha=0.7, edgecolor='black', log=True)
    plt.title(f"Weights histogram: {name}")
    plt.grid(True)
    plt.show()
```
